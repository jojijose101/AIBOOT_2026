```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Simulate 3D data
np.random.seed(0)
n_samples = 100
X = np.random.rand(n_samples, 3) * 10

# Center the data
X_centered = X - np.mean(X, axis=0)

# Covariance matrix
cov_matrix = np.cov(X_centered.T)

# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvectors by decreasing eigenvalues
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Visualize the centered 3D data with eigenvectors
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_centered[:, 0], X_centered[:, 1], X_centered[:, 2], alpha=0.6)

# Draw global coordinate axes
axis_len = 8
ax.plot([-axis_len, axis_len], [0, 0], [0, 0], color='r', linewidth=2, label='X-axis')
ax.plot([0, 0], [-axis_len, axis_len], [0, 0], color='g', linewidth=2, label='Y-axis')
ax.plot([0, 0], [0, 0], [-axis_len, axis_len], color='b', linewidth=2, label='Z-axis')

# Draw eigenvectors (principal components)
origin = np.array([0, 0, 0])
for i in range(3):
    ax.quiver(*origin,
              *(eigenvectors[:, i] * np.sqrt(eigenvalues[i]) * 3),  # scaled for visibility
              color=f'C{i+3}', linewidth=2, arrow_length_ratio=0.1,
              label=f'PC{i+1}')

# Labels & limits
ax.set_title('3D Data (Centered) with Global Axes and PCA Axes')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])

ax.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Simulate 3D data
np.random.seed(0)
n_samples = 100
X = np.random.rand(n_samples, 3) * 10

# Center the data
X_centered = X - np.mean(X, axis=0)

# Covariance matrix & eigen decomposition
cov_matrix = np.cov(X_centered.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Transformation matrix for ellipsoid
transform_matrix = eigenvectors @ np.diag(np.sqrt(eigenvalues))

# Create a unit sphere
u = np.linspace(0, 2 * np.pi, 50)
v = np.linspace(0, np.pi, 50)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

# Transform sphere to ellipsoid
ellipsoid = np.dot(np.c_[x.flatten(), y.flatten(), z.flatten()], transform_matrix)
ellipsoid = ellipsoid.reshape(x.shape[0], x.shape[1], 3)

# Visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of centered data
ax.scatter(X_centered[:, 0], X_centered[:, 1], X_centered[:, 2], alpha=0.6)

# Draw global coordinate axes
axis_len = 8
ax.plot([-axis_len, axis_len], [0, 0], [0, 0], color='r', linewidth=2, label='Global X-axis')
ax.plot([0, 0], [-axis_len, axis_len], [0, 0], color='g', linewidth=2, label='Global Y-axis')
ax.plot([0, 0], [0, 0], [-axis_len, axis_len], color='b', linewidth=2, label='Global Z-axis')

# Draw PCA eigenvectors
origin = np.array([0, 0, 0])
for i in range(3):
    ax.quiver(*origin,
              *(eigenvectors[:, i] * np.sqrt(eigenvalues[i]) * 3),
              color=f'C{i+3}', linewidth=2, arrow_length_ratio=0.1,
              label=f'PC{i+1}')

# Draw covariance ellipsoid
ax.plot_surface(ellipsoid[:, :, 0], ellipsoid[:, :, 1], ellipsoid[:, :, 2],
                rstride=4, cstride=4, color='c', alpha=0.3)

# Labels & limits
ax.set_title('3D Data (Centered) with Global Axes, PCA Axes, and Covariance Ellipsoid')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])

ax.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Simulate some 3D data
np.random.seed(0)
n_samples = 100
X = np.random.rand(n_samples, 3) * 10

# Center the data
X_centered = X - np.mean(X, axis=0)

# Covariance & PCA
cov_matrix = np.cov(X_centered.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Project data onto PC1
X_projected_on_pc1 = X_centered @ eigenvectors[:, 0]
projected_points_3d = np.outer(X_projected_on_pc1, eigenvectors[:, 0])

# Visualization
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Projected data
ax.scatter(projected_points_3d[:, 0], projected_points_3d[:, 1], projected_points_3d[:, 2],
           alpha=0.6, label='Projected Data (on PC1)')

# Draw PC1 vector
origin = np.array([0, 0, 0])
scale = np.max(np.abs(X_projected_on_pc1)) * 1.5
ax.quiver(*origin,
          *(eigenvectors[:, 0] * scale),
          color='red', linewidth=2, arrow_length_ratio=0.1,
          label='PC1')

# Draw global coordinate axes
axis_len = scale
ax.plot([-axis_len, axis_len], [0, 0], [0, 0], color='black', linewidth=1.5, label='X-axis')
ax.plot([0, 0], [-axis_len, axis_len], [0, 0], color='green', linewidth=1.5, label='Y-axis')
ax.plot([0, 0], [0, 0], [-axis_len, axis_len], color='blue', linewidth=1.5, label='Z-axis')

# Labels & limits
ax.set_title('3D Data Projected onto PC1 with Global Axes')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-axis_len, axis_len])
ax.set_ylim([-axis_len, axis_len])
ax.set_zlim([-axis_len, axis_len])

ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Project the centered data onto the first two principal components (PC1 and PC2)
X_projected_on_pc12 = X_centered @ eigenvectors[:, :2]

# Reconstruct projected points in the original 3D space
projected_points_3d_pc12 = X_projected_on_pc12 @ eigenvectors[:, :2].T

# Visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of projected points
ax.scatter(projected_points_3d_pc12[:, 0],
           projected_points_3d_pc12[:, 1],
           projected_points_3d_pc12[:, 2],
           alpha=0.6, label='Projected Data (on PC1 & PC2)')

# Eigenvector scaling
origin = np.array([0, 0, 0])
scale_pc1 = np.max(np.abs(X_projected_on_pc12[:, 0]))
scale_pc2 = np.max(np.abs(X_projected_on_pc12[:, 1]))

# Draw PC1 and PC2 vectors
ax.quiver(*origin,
          *(eigenvectors[:, 0] * scale_pc1 * 1.2),
          color='red', linewidth=2, arrow_length_ratio=0.1, label='PC1')

ax.quiver(*origin,
          *(eigenvectors[:, 1] * scale_pc2 * 1.2),
          color='purple', linewidth=2, arrow_length_ratio=0.1, label='PC2')

# Draw global coordinate axes
axis_len = max(scale_pc1, scale_pc2) * 1.5
ax.plot([-axis_len, axis_len], [0, 0], [0, 0], color='black', linewidth=1.5, label='X-axis')
ax.plot([0, 0], [-axis_len, axis_len], [0, 0], color='green', linewidth=1.5, label='Y-axis')
ax.plot([0, 0], [0, 0], [-axis_len, axis_len], color='blue', linewidth=1.5, label='Z-axis')

# Draw translucent PC1-PC2 plane
grid_size = 10
pc1_vals = np.linspace(-scale_pc1, scale_pc1, grid_size)
pc2_vals = np.linspace(-scale_pc2, scale_pc2, grid_size)
PC1_grid, PC2_grid = np.meshgrid(pc1_vals, pc2_vals)
plane_points = PC1_grid[..., None] * eigenvectors[:, 0] + PC2_grid[..., None] * eigenvectors[:, 1]
ax.plot_surface(plane_points[..., 0], plane_points[..., 1], plane_points[..., 2],
                color='orange', alpha=0.2, rstride=1, cstride=1)

# Labels & limits
ax.set_title('3D Data Projected onto Plane of PC1 and PC2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-axis_len, axis_len])
ax.set_ylim([-axis_len, axis_len])
ax.set_zlim([-axis_len, axis_len])

ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Project the centered data onto the first two principal components (PC1 and PC2)
X_projected_on_pc12 = X_centered @ eigenvectors[:, :2]

# Reconstruct projected points in the original 3D space
projected_points_3d_pc12 = X_projected_on_pc12 @ eigenvectors[:, :2].T

# Visualization
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of projected points
ax.scatter(projected_points_3d_pc12[:, 0],
           projected_points_3d_pc12[:, 1],
           projected_points_3d_pc12[:, 2],
           alpha=0.6, label='Projected Data (on PC1 & PC2)')

# Eigenvector scaling
origin = np.array([0, 0, 0])
scale_pc1 = np.max(np.abs(X_projected_on_pc12[:, 0]))
scale_pc2 = np.max(np.abs(X_projected_on_pc12[:, 1]))
axis_len = max(scale_pc1, scale_pc2) * 1.5  # for drawing axes

# Draw PC1 and PC2 vectors
ax.quiver(*origin,
          *(eigenvectors[:, 0] * scale_pc1 * 1.2),
          color='red', linewidth=2, arrow_length_ratio=0.1, label='PC1')

ax.quiver(*origin,
          *(eigenvectors[:, 1] * scale_pc2 * 1.2),
          color='purple', linewidth=2, arrow_length_ratio=0.1, label='PC2')

# Draw translucent PC1-PC2 plane
grid_size = 10
pc1_vals = np.linspace(-scale_pc1, scale_pc1, grid_size)
pc2_vals = np.linspace(-scale_pc2, scale_pc2, grid_size)
PC1_grid, PC2_grid = np.meshgrid(pc1_vals, pc2_vals)
plane_points = PC1_grid[..., None] * eigenvectors[:, 0] + PC2_grid[..., None] * eigenvectors[:, 1]
ax.plot_surface(plane_points[..., 0], plane_points[..., 1], plane_points[..., 2],
                color='orange', alpha=0.2, rstride=1, cstride=1)

# ---- Draw XYZ axes ----
ax.quiver(*origin, axis_len, 0, 0, color='blue', linewidth=1.5, arrow_length_ratio=0.05, label='X-axis')
ax.quiver(*origin, 0, axis_len, 0, color='green', linewidth=1.5, arrow_length_ratio=0.05, label='Y-axis')
ax.quiver(*origin, 0, 0, axis_len, color='black', linewidth=1.5, arrow_length_ratio=0.05, label='Z-axis')

# Labels & limits
ax.set_title('3D Data Projected onto Plane of PC1 and PC2 (with XYZ axes)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-axis_len, axis_len])
ax.set_ylim([-axis_len, axis_len])
ax.set_zlim([-axis_len, axis_len])

# ---- Set view perpendicular to PC1-PC2 plane (looking along PC3) ----
pc3 = eigenvectors[:, 2]
elev = np.degrees(np.arcsin(pc3[2]))   # vertical angle from Z-axis
azim = np.degrees(np.arctan2(pc3[1], pc3[0]))  # horizontal rotation
ax.view_init(elev=elev, azim=azim)

ax.legend()
plt.tight_layout()
plt.show()
```

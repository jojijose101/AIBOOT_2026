## **Gradient Descent**

```python
def f(p):
  return (p[0]-2)**2+(p[1]+3)**2
```

The function $f(x,y)$ is defined as:

$$f(x,y) = (x - 2)^2 + (y + 3)^2$$

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and an axes object for the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create a grid of points
x = np.linspace(-15, 15, 100)
y = np.linspace(-15, 15, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z values using the function f
Z = np.array([[f((xi, yi)) for xi in x] for yi in y])

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Add labels and title
ax.set_xlabel('p[0]')
ax.set_ylabel('p[1]')
ax.set_zlabel('f(p)')
ax.set_title('3D Surface Plot of f(p)')

plt.show()
```

```python
def gradf(p):
  return np.array((2*p[0]-4, 2*p[1]+6))
```

The gradient function $\nabla f(x,y)$ is defined as:

$$\nabla f(x,y) = (2x - 4) + (2y + 6)$$

```python
import numpy as np
def gradent_decent(p0,f,gradf,alpha,max_itr=100):
  p=np.array(p0)
  path = [p]
  bound = np.array((0.001,0.001))
  i=0
  while(not np.all(gradf(p)<bound) and i<=max_itr):
    # print(p) # Commenting out print for cleaner output, will use plot instead
    p=p-alpha*gradf(p)
    path.append(p)
    i=i+1
  return p, np.array(path)
```

```python
solution = gradent_decent((-10,10),f,gradf,0.1)
print(np.round(solution[0]))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and an axes object for the 3D plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Create a grid of points for the surface
x_surface = np.linspace(-15, 15, 100)
y_surface = np.linspace(-15, 15, 100)
X_surface, Y_surface = np.meshgrid(x_surface, y_surface)

# Calculate Z values for the surface using the function f
Z_surface = np.array([[f((xi, yi)) for xi in x_surface] for yi in y_surface])

# Plot the surface with transparency
ax.plot_surface(X_surface, Y_surface, Z_surface, cmap='viridis', alpha=0.6, rstride=5, cstride=5)

# Get the gradient descent path and its corresponding Z values
path = solution[1]
path_z = np.array([f(p_val) for p_val in path])

# Plot the gradient descent path with a thick line and markers
ax.plot(path[:, 0], path[:, 1], path_z, 'r-o', markersize=6, linewidth=3, label='Gradient Descent Path')

# Add distinct start and end markers for the path
ax.plot([path[0, 0]], [path[0, 1]], [path_z[0]], 'go', markersize=12, label='Start Point')
ax.plot([path[-1, 0]], [path[-1, 1]], [path_z[-1]], 'X', color='black', markersize=15, linewidth=4, label='End Point')

# Add labels and title
ax.set_xlabel('p[0]')
ax.set_ylabel('p[1]')
ax.set_zlabel('f(p)')
ax.set_title('3D Surface Plot of f(p) with Gradient Descent Path')
ax.legend()

plt.show()
```
